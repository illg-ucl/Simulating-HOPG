```python
%matplotlib notebook
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
    """Egg-Carton Potential Function
    x = 1D NumPy array
    y = 1D NumPy array
    """
    return np.sin(x)+np.cos(y)
```

```python
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)

x_mesh,y_mesh = np.meshgrid(x,y)

z = f(x_mesh,y_mesh)
```

```python
fig = plt.figure(figsize=[8,8])
ax = plt.axes(projection="3d")

ax.plot_surface(x_mesh,y_mesh,z,cmap="inferno")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.savefig("carton1.png")
```
